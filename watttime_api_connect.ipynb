# WattTime API Demo
First, register for an account with an HTTP POST to WattTime's API:

```python
import urllib3

http = urllib3.PoolManager()

values = """
  {
    "username": "journalclub",
    "password": "weloveSEAC",
    "email": "matthew.irish@nrel.gov",
    "org": "NREL"
  }
"""

headers = {
  'Content-Type': 'application/json'
}


r = http.request('POST','https://api2.watttime.org/v2/register', body=values, headers=headers)

print(r.data)
```

Now login and get an access token that we can then use to get the data!

```python
import base64

http = urllib3.PoolManager()

auth_string = "journalclub:weloveSEAC".encode('ascii')
auth_string = base64.b64encode(auth_string) 

headers = {
  'Authorization': auth_string
}


r = http.request('POST','https://api2.watttime.org/v2/login', headers=headers)

print(r.data)
```
